we have 928 different categories and among data 34 of them are missing. it's take a lot of time to find the appropriate categories for them so we're going to drop them. 

```python
df1.dropna(subset=["categories"], axis=0, inplace=True)
```

### time to change data type

```python
df1[["name"]]=df1[["name"]].astype("str")
df1[["address"]]=df1[["address"]].astype("str")
df1[["city"]]=df1[["city"]].astype("str")
df1[["state"]]=df1[["state"]].astype("str")
#postal code in canada are a combinatian of numbers and string
df1[["postal_code"]]=df1[["postal_code"]].astype("str")
df1[["stars"]]=df1[["stars"]].astype("float")
df1[["categories"]]=df1[["categories"]].astype("str")
df1.dtypes
```

Now we remove any whitespace in names,address,city, and state and also change all letters to lowercase

```python
df1.name=df1.name.str.strip().str.lower()
df1.address=df1.address.str.strip().str.lower()
df1.city=df1.city.str.strip().str.lower()
df1.state=df1.state.str.strip().str.lower()
df1.categories=df1.categories.str.strip().str.lower()
```
